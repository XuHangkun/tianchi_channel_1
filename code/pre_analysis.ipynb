```python
train_df = pd.read_csv(os.path.join(proj_top,'tcdata/medical_nlp_round1_data/train.csv'),sep="\|,\|",names=["id","report","label"],index_col=0)
```

## Report Length distribution

```python
def tokenize(text):
    try:
        return [int(x) for x in text.split()]
    except:
        return []
length = []
labels = []
for index in range(len(train_df)):
    report = train_df["report"][index]
    length.append(len(tokenize(report)))
    label = train_df["label"][index]
    labels += tokenize(label)
```

```python
plt.hist(length, bins=100, facecolor="blue", alpha=0.7)
plt.xlabel("Length of report")
plt.ylabel("Count")
```

```python
plt.hist(labels,range=[0,17],bins=17, facecolor="blue", alpha=0.7)
plt.xlabel("Labels")
plt.ylabel("Count")
```
